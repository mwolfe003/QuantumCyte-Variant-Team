## replacing values with NA as 0

```python
df_final =pd.read_excel('/content/Final data for Bidhan 4.4.23 _JB Tumor Cellularity Plot.xlsx')
df_final
```

```python
df_final[['Qcyte AF', 'Manual VAF']] = df_final[['Qcyte AF', 'Manual VAF']].fillna(0)
```

```python
df_final['VAF increase'] = df_final['Qcyte AF'] - df_final['Manual VAF']
df_final['VAF increase proportion'] = df_final['Qcyte AF'] / df_final['Manual VAF']
```

```python
df_final['VAF increase proportion'].value_counts(dropna= False)
```

```python
df_final[['Manual VAF', 'Qcyte AF', 'VAF increase proportion']]
```

```python
print('median', df_final['VAF increase'].median())
print('mean', df_final['VAF increase'].mean())
print('max', df_final['VAF increase'].max())
print('min', df_final['VAF increase'].min())
```

```python
df_final['VAF increase proportion']
```

```python
df_final['VAF increase proportion'].value_counts()
```

```python
print('median', df_final['VAF increase proportion'].median())
print('mean', df_final['VAF increase proportion'].mean())
print('max', df_final['VAF increase proportion'].max())
print('min', df_final['VAF increase proportion'].min())
```

```python
from scipy.stats import ttest_rel

# Drop rows with NaNs in either column to avoid errors
paired_data = df_final[['Manual VAF', 'Qcyte AF']].dropna()

# Perform paired t-test
statistic, p_value = ttest_rel(paired_data['Qcyte AF'], paired_data['Manual VAF'])

print(f"t-statistic: {statistic}")
print(f"p-value: {p_value}")
```

```python
from scipy.stats import wilcoxon

# Drop NaNs
paired_data = df_final[['Manual VAF', 'Qcyte AF']].dropna()

# Wilcoxon test
statistic, p_value = wilcoxon(paired_data['Qcyte AF'], paired_data['Manual VAF'])

print(f"Wilcoxon statistic: {statistic}")
print(f"p-value: {p_value}")
```
